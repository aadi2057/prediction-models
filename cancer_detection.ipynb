```python
# Splitting the independent (X) and the dependent (y) data sets 
X = df.iloc[:, 2:].values
y = df.iloc[:, 1].values
```

```python
# Splitting the data into training and the test datasets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
y_test.shape
```

```python
# Scale the data (feature scaling)
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Create funcitons for models
def models(X_train, y_train, X_test, y_test):
    """This funciton builds different models for classifying the Breast Cancer."""
#     Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, y_train)
    
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    tree.fit(X_train, y_train)
    
    #Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state=0)
    forest.fit(X_train, y_train)
    
    #Print the accuracies of the models
    print('Accuracy of Logistic Regression: Train = {} Test= {}'.format(log.score(X_train, y_train), log.score(X_test, y_test)))
    print('Accuracy of Decision Tree: Train = {} Test= {}'.format(tree.score(X_train, y_train), tree.score(X_test, y_test)))
    print('Accuracy of Random Forest: Train = {} Test= {}'.format(forest.score(X_train, y_train), forest.score(X_test, y_test)))
    
    return log, tree, forest
```

```python
model = models(X_train, y_train, X_test, y_test)
```

```python
# Test data accuracy on confusion Matrix (Alternative Method.)
from sklearn.metrics import confusion_matrix

for i in range(len(model)) :
    print('Model: ', i)
    
    cm = confusion_matrix(y_test, model[i].predict(X_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[0][1]
    FN = cm[1][0]
    
    print('Testing Accuracy= ', (TP + TN) / (TP + TN + FP + FN))
    print()

    
```

```python
# Another way for test accuracy of the models
from sklearn.metrics import classification_report, accuracy_score

for i in range(len(model)):
    print('Model: ', i)
    print(classification_report(y_test, model[i].predict(X_test)))
    print(accuracy_score(y_test, model[i].predict(X_test)))
    print()
```
